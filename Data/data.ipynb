```python
country_codes ={}
with open("codes.txt","r") as file:
    countries = file.readlines()
    for country in range(0,len(countries)):
        country_list = countries[country].split("\t")

        country_codes[country_list[0]] = f"+{country_list[1]}"


print(country_codes)
```

```python
import json

with open("Country.json","w+") as js_file:
    json.dump(country_codes,js_file)
```
